# 5. Feature Engineering

```python
customer_data_altered = customer_data.copy()

# Removing the ID column
customer_data_altered = customer_data_altered.drop(columns = 'ID')

# Dropping rows which contain outliers

customer_data_altered = customer_data_altered.drop(index = index_outliers)

# Transforming the Education into an ordinal integer
educ_mapper = {'Basic' : 1, 'Graduation' : 1 ,'2n Cycle' : 2, 'Master' : 2, 'PhD' : 3}
customer_data_altered['Education'] = customer_data_altered['Education'].replace(educ_mapper)

# Number of kids/children at home
customer_data_altered['number_of_kids_child_home'] = customer_data_altered.Kidhome + customer_data_altered.Teenhome

# Number of purchases across all channels
customer_data_altered['number_purchases'] = customer_data_altered.NumWebPurchases + customer_data_altered.NumCatalogPurchases + customer_data_altered.NumStorePurchases

# Transforming Marital Status by combining low occurence categories into 'Other'
customer_data_altered['Marital_Status'] = np.where(customer_data_altered.Marital_Status.isin(['Alone','YOLO','Absurd']),'Other', customer_data_altered.Marital_Status)

# Imputation by mean
customer_data_altered['Income'] = np.where(customer_data_altered.Income.isnull(),customer_data_altered.Income.mean(), customer_data_altered.Income)


# Retrieving the age of the customer (Base year 2022)
customer_data_altered['customer_age'] = 2022 - customer_data_altered.Year_Birth


# Dropping Maritual Status and Dt_Customer for now
customer_data_altered = customer_data_altered.drop(columns = ['Marital_Status','Dt_Customer','Year_Birth'])

# Dropping Maritual Status and Dt_Customer for now
customer_data_altered['total_spend'] =  customer_data_altered['MntWines'] +  customer_data_altered['MntFruits'] +  customer_data_altered['MntMeatProducts'] +  customer_data_altered['MntFishProducts'] +  customer_data_altered['MntSweetProducts'] +  customer_data_altered['MntGoldProds']
```

# 6. Clustering

## 6.1 K-means

```python
std_scaler = StandardScaler()
customer_data_standardized = std_scaler.fit_transform(customer_data_altered)

# Fitting a PCA object in order to visualize the clusters
pca_2 = PCA(n_components = 2)
pca_2 = pca_2.fit(customer_data_standardized)
```

```python
# Explained variance of each of the eigenvectors
explained_variance = pca_2.explained_variance_
plt.bar(['1','2'],explained_variance)
```

```python
# Plotting intertia in order to select correct k-value for K-means
distortions = list()
k_value = range(1,10,1)
for k in k_value:
    k_means = KMeans(n_clusters=k)
    k_means.fit(customer_data_altered)
    distortions.append(k_means.inertia_)
```

```python
plt.plot(k_value, distortions)
```

```python
# Initializing a K-means object
k_means = KMeans(n_clusters=3)

# Fitting the K-means to the customer data 
k_means_fitted = k_means.fit(customer_data_standardized)
```

```python
k_means_fitted.labels_
```

```python
pca_2_values = pca_2.transform(customer_data_standardized)

customer_data__fitted = customer_data_altered.copy() 
customer_data__fitted['fitted_labels'] = k_means.labels_
customer_data__fitted['pca_1'] = pca_2_values[:,0]
customer_data__fitted['pca_2'] = pca_2_values[:,1]
```

```python
customer_data__fitted.fitted_labels.value_counts().plot(kind = 'bar')
```
