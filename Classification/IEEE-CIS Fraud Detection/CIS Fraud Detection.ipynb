### Transaction Table 
**TransactionDT**: timedelta from a given reference datetime (not an actual timestamp) \
**TransactionAMT**: transaction payment amount in USD \
**ProductCD**: product code, the product for each transaction \
**card1 - card6**: payment card information, such as card type, card category, issue bank, country, etc. \
**addr: address** \
**dist: distance** \
**P_ and (R__) emaildomain:** purchaser and recipient email domain \
**C1-C14**: counting, such as how many addresses are found to be associated with the payment card, etc. The actual meaning is masked. \
**D1-D15**: timedelta, such as days between previous transaction, etc. \
**M1-M9**: match, such as names on card and address, etc. \
**Vxxx**: Vesta engineered rich features, including ranking, counting, and other entity relations. \


### Identity Table
Variables in this table are identity information – network connection information (IP, ISP, Proxy, etc) and digital signature (UA/browser/os/version, etc) associated with transactions.
They're collected by Vesta’s fraud protection system and digital security partners.
(The field names are masked and pairwise dictionary will not be provided for privacy protection and contract agreement)

# 5. Feature Engineering


```python
modelling_dataset = primary_train_dataset[relevant_features]
modelling_dataset['isFraud'] = primary_train_dataset.isFraud
```

```python
# Handling missing values
for g in relevant_features:
    feature_dtype = relevant_features_dataset[g].dtypes
    if feature_dtype == 'object':
        modelling_dataset[g] = modelling_dataset[g].replace(np.NaN,'missing')
    elif feature_dtype in ('float64','int64'):
        modelling_dataset[g] = modelling_dataset[g].replace(np.NaN,modelling_dataset[g].mean())
    else:
        print('Unknown Dtype')
```

```python
modelling_dataset['P_emaildomain'] = np.where(~modelling_dataset['P_emaildomain'].isin(['protonmail','mail.com','outlook.es','aim.com','outlook.com']),'Other',modelling_dataset['P_emaildomain'])
modelling_dataset['R_emaildomain'] = np.where(~modelling_dataset['R_emaildomain'].isin(['protonmail','mail.com','netzero.net','outlook.com','outlook.es','icloud.com','gmail.com']),'Other',modelling_dataset['R_emaildomain'])                                               


columns_high_fraud_ratio_id_30 = id_30_fraud_ratio[id_30_fraud_ratio.isFraud > 0.10].index.tolist()
modelling_dataset['id_30'] = np.where(modelling_dataset['id_30'].isin(columns_high_fraud_ratio_id_30),modelling_dataset['id_30'],'Other')

columns_high_fraud_ratio_id_31 = id_31_fraud_ratio[id_31_fraud_ratio.isFraud > 0.10].index.tolist()
modelling_dataset['id_31'] = np.where(modelling_dataset['id_31'].isin(columns_high_fraud_ratio_id_31),modelling_dataset['id_31'],'Other')

columns_high_fraud_ratio_id_33 = id_33_fraud_ratio[id_33_fraud_ratio.isFraud > 0.10].index.tolist()
modelling_dataset['id_33'] = np.where(modelling_dataset['id_33'].isin(columns_high_fraud_ratio_id_33),modelling_dataset['id_33'],'Other')

                                                                      
high_fraud_rate_col_selected = modelling_dataset[modelling_dataset.DeviceInfo.isin(high_fraud_count_col)].groupby('DeviceInfo').mean().sort_values(by='isFraud', ascending = False).head(25)
high_fraud_rate_col_final = high_fraud_rate_col_selected[high_fraud_rate_col_selected.isFraud > 0.20].index.tolist()
modelling_dataset['DeviceInfo'] = np.where(modelling_dataset['DeviceInfo'].isin(high_fraud_rate_col_final),modelling_dataset['DeviceInfo'],'Other')
                                                                            
                                                     
```

We can see that the correlation between isFraud and the quantitative variables fell after imputation.

```python
# Extracting features where a hot encoding should be done.
relevant_dummy_variables = list()
for g in modelling_dataset.columns:
    if modelling_dataset[g].dtypes == 'object':
        relevant_dummy_variables.append(g)
        
    
```

```python
# Creating dummy variables
dummy_variables = pd.get_dummies(modelling_dataset[relevant_dummy_variables])
modelling_dataset[dummy_variables.columns.tolist()] = dummy_variables
modelling_dataset = modelling_dataset.drop(columns = relevant_dummy_variables)
```

# 6. Modelling

## 6.1 Preprocessing

```python
target_variables = modelling_dataset.isFraud
feature_variables = modelling_dataset.drop(columns = 'isFraud')
numerical_features = feature_variables.select_dtypes(include = ['float','int']).columns.tolist()

# Scaling the numerical features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(feature_variables[numerical_features])
scaled_features_df = pd.DataFrame(scaled_features)
scaled_features_df.columns = numerical_features
feature_variables = feature_variables.drop(columns = numerical_features)
feature_variables[numerical_features] = scaled_features_df


# Train splitting the dataset
x_train, x_validation, y_train, y_validation = train_test_split(feature_variables,target_variables, test_size = 0.20, random_state = 0)

```

```python
training_set = x_train
training_set['isFraud'] = y_train

majority_class_dataset = training_set[training_set.isFraud == 0]
minority_class_dataset = training_set[training_set.isFraud == 1]
majority_class_dataset = majority_class_dataset.sample(frac = 0.20, random_state = 0)

training_set_undersampled = pd.concat([majority_class_dataset,minority_class_dataset])
y_train = training_set_undersampled.isFraud
x_train = training_set_undersampled.drop(columns = 'isFraud')
```

```python
len(y_train[y_train == 1])/(len(y_train))
```

## 6.2 Model Training

```python
# Initializing a Random Forest Model
random_forest = RandomForestClassifier(n_estimators = 100)

# Fitting the Random Forest to our training data.
random_forest = random_forest.fit(x_train,y_train)
```

```python
# Initializing a Gradient Boost Model
gradient_boosted_trees = GradientBoostingClassifier()

# Fitting the Gradient Boost to our training data.
gradient_boosted_trees = gradient_boosted_trees.fit(x_train,y_train)
```

```python
# Initializing a Xtreme Gradient Boosted Model
xg_boost = xgb.XGBClassifier(
                max_depth = 10, alpha = 5, n_estimators = 500)
xg_boost.fit(x_train,y_train)
```

```python
# Initializing a Logistic Regression
logistic_regression = LogisticRegression(max_iter = 500)
logistic_regression.fit(x_train,y_train)
```

## 6.3 Model Evaluation ( Validation Dataset)

```python
# Predicting the validation dataset for each model
predicted_fraud_validation_rf = random_forest.predict_proba(x_validation)
predicted_fraud_validation_gbs = gradient_boosted_trees.predict_proba(x_validation)
predicted_fraud_validation_xgb = xg_boost.predict_proba(x_validation)
predicted_fraud_validation_lr = logistic_regression.predict_proba(x_validation)
```

```python
rf_auc_score_validation = roc_auc_score(y_validation.values,predicted_fraud_validation_rf[:,1])
gbs_auc_score_validation = roc_auc_score(y_validation.values,predicted_fraud_validation_gbs[:,1])
xgb_auc_score_validation = roc_auc_score(y_validation.values,predicted_fraud_validation_xgb[:,1])
lr_auc_score_validation = roc_auc_score(y_validation.values,predicted_fraud_validation_lr[:,1])
print(f"RF AUC : {rf_auc_score_validation}\nGBS AUC: {gbs_auc_score_validation}\nXGB AUC: {xgb_auc_score_validation}\nLR AUC: {lr_auc_score_validation}")
```

```python
rf_pr_score_validation = average_precision_score(y_validation.values,predicted_fraud_validation_rf[:,1])
gbs_pr_score_validation = average_precision_score(y_validation.values,predicted_fraud_validation_gbs[:,1])
xgb_pr_score_validation = average_precision_score(y_validation.values,predicted_fraud_validation_xgb[:,1])
lr_pr_score_validation = average_precision_score(y_validation.values,predicted_fraud_validation_lr[:,1])
print(f"RF PR : {rf_pr_score_validation}\nGBS PR: {gbs_pr_score_validation}\nXGB PR: {xgb_pr_score_validation}\nLR PR: {lr_pr_score_validation}")
```

```python
precision, recall,_ = precision_recall_curve(y_validation.values,predicted_fraud_validation_xgb[:,1])
```

```python
disp = PrecisionRecallDisplay(precision=precision, recall=recall)
```

## 6.4 Final Model

```python
feature_variables_final = feature_variables.copy()
feature_variables_final['isFraud'] = target_variables
majority_class_dataset_final = feature_variables_final[feature_variables_final.isFraud == 0]
minority_class_dataset_final = feature_variables_final[feature_variables_final.isFraud == 1]
majority_class_dataset_final = majority_class_dataset_final.sample(frac = 0.20, random_state = 0)

training_set_undersampled_final = pd.concat([majority_class_dataset_final,minority_class_dataset_final])
y_train_final = training_set_undersampled_final.isFraud
x_train_final = training_set_undersampled_final.drop(columns = 'isFraud')
```

```python
# Define the search space
param_grid = { 
    # Learning rate shrinks the weights to make the boosting process more conservative
    "learning_rate": [0.0001,0.001, 0.01, 0.1, 1] ,
    # Maximum depth of the tree, increasing it increases the model complexity.
    "max_depth": range(3,21,3),
    # Gamma specifies the minimum loss reduction required to make a split.
    "gamma": [i/10.0 for i in range(0,5)],
    # Percentage of columns to be randomly samples for each tree.
    "colsample_bytree": [i/10.0 for i in range(3,10)],
    # reg_alpha provides l1 regularization to the weight, higher values result in more conservative models
    "reg_alpha": [1e-5, 1e-2, 0.1, 1, 10, 100],
    # reg_lambda provides l2 regularization to the weight, higher values result in more conservative models
    "reg_lambda": [1e-5, 1e-2, 0.1, 1, 10, 100],
    
    "n_estimators" : [100,200,300,500,1000]

}
# Set up score
scoring = ['roc_auc']
```

```python
# Define random search
random_search = RandomizedSearchCV(estimator=xg_boost, 
                           param_distributions=param_grid, 
                           n_iter=4,
                           scoring=scoring, 
                           refit='recall', 
                           n_jobs=-1, 
                           cv=2, 
                           verbose=0)
# Fit RandomSearchCV search
random_result = random_search.fit(x_train_final, y_train_final)
# Print grid search summary
random_result
# Print the best score and the corresponding hyperparameters
print(f'The best score is {random_result.best_score_:.4f}')
print(f'The best hyperparameters are {random_result.best_params_}')
```

```python
xg_boost_final = xgb.XGBClassifier(
                max_depth = 10,
                reg_lambda = 0.1,
                reg_alpha = 0.0001,
                learning_rate = 1, 
                gamma = 0.1,
                colsample_bytree =  0.6,
                n_estimators = 500,
                random_state = 0)

xg_boost_final.fit(x_train_final,y_train_final)
```
